```python
#filtering the columns: 'Order ID', ‘Customer Name’,’Country’,‘Category	’,'Product Name','Region', and ‘Sales’ from the file
filtered_data = data[['Order Date','Order ID', 'Customer Name', 'State','Region','Country','Sub-Category', 'Category', 'Product Name','Sales']]
filtered_data.head(2)
```

```python
import matplotlib.pyplot as plt
```

```python
#Creating a pie chart with percentages category wise 
filtered_data["Category"].value_counts().plot(kind='pie',autopct='%1.1f%%', startangle=90,shadow = True)
plt.ylabel('')
plt.title("Product Category Sales")
plt.show()
```

**1. Sales & Profit Analysis**

```python
#What are the top 5 product categories with the highest total sales?
```

```python
categorysales = filtered_data.groupby('Sub-Category')['Sales'].sum().reset_index()
top_category_product = categorysales.sort_values(by='Sales', ascending = False).head(5)
top_category_product
plt.figure(figsize=(10,6))
plt.bar(top_category_product['Sub-Category'],top_category_product['Sales'],color = "orange")
plt.title("Top 5 product Sub-Category with Highest Sales")
plt.xlabel("Sub-Category")
plt.ylabel("Total Sales")
plt.show()
```

Which region has the highest sales margin and which has the lowest?

```python
region_sales = filtered_data.groupby('Region')['Sales'].sum().reset_index()
highest_region_sales = region_sales.loc[region_sales.idxmax()]
lowest_region_sales = region_sales.loc[region_sales.idxmin()]
print(f'Highest sale:{highest_region_sales}')
print(f'Lowest sale:{lowest_region_sales}')
```

Which product sub-category generates the highest sales?

```python
#we can group the data by Sub-Category and calculate the total sales for each one
import seaborn as sns
top_sub_categorysales = categorysales.sort_values(by='Sales', ascending = False).head(10)
#plotting using seaborn
plt.figure(figsize=(10,6))
sns.barplot(x='Sales',y='Sub-Category',data = top_sub_categorysales)
plt.show()
```
